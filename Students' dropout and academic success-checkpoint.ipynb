```python
import matplotlib as mpl
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
import pandas as pd
import numpy as np 
import statsmodels.api as sm
sns.set_style("darkgrid")
mpl.rcParams['figure.figsize'] = (20,5)
```

```python
df=pd.read_csv("Student dropout & academic success.csv")
df.head()
```

# Exploratory Data Analysis


### 1) Creating new columns

```python
# Adding two columns to the dataset: "Dropout" and "Graduate" conlumns to perform correlation functions and 
# other numerical operations.

df['Dropout']= df['Student Status'].apply(lambda x:1 if x== 'Dropout'else 0 if x != 'Enrolled' else np.nan)
df
```

```python
df['Graduate']=df['Student Status'].apply(lambda x:1 if x== 'Graduate'else 0 if x != 'Enrolled'else np.nan)
df
```

### 2) Correlation Analysis (Dropout)
Given that the main focus of this analysis is the factors influencing student dropout or academic success, an analysis identifying these factors will frame the perspective of how to start the analysis. 

```python
# Correlation between all other variables and 'Dropout'column. The 'Student Status' column is excludend from this operation.

column_to_exclude= ['Student Status']
correlation_result= df.drop(columns=column_to_exclude).corr()['Dropout']
correlation_result.sort_values(ascending=False)
```

### Correlation Analysis Dropout Heatmap
For better represenation of our correlation to Dropout variable results, we create a heatmap.


**Insights**: The following are results following a correlation analysis of all variables against the 'Dropout' variable:

-  Age at enrollment,Debtor, and Gender have the highest indexes of correlation with the Dropout variable.
-  The correlation indexes range from .25 to .27. Overall these indexes are not really strong indicators of correlation, but they are worth persuing futher analysis and delve deeper into understanding underlying relationships that can provide further insights.

```python
sorted_correlation_result = correlation_result.sort_values(ascending=False)
plt.figure(figsize=(10,8))
sns.heatmap(sorted_correlation_result.to_frame(), annot=True,cmap="rocket_r",fmt=".2f")
plt.show()
```

### 3) Correlation Analysis (Graduate)

```python
# Correlation between all other variables and 'Graduate'column. The 'Student Status' column is excludend from this operation.
column_to_exclude= ['Student Status']
correlation_result= df.drop(columns=column_to_exclude).corr()['Graduate']
correlation_result.sort_values(ascending=False)
```

### Correlation Analysis Graduate Heatmap

**Insights**: The following results following a correlation analysis from all variables against the 'Graduate' variable:

-  Curricular units 2nd sem(approved),Curricular units 1st sem (approved),Curricular units 2nd sem(grade),Curricular units 1st sem(grade),and Tuition fees up to date have the highest indexes of correlation with Graduate variable.
-  The indexes range from as low as .44 to as high as .65. Overall these indexes are moderately strong indicators of correlation. Next, a deeper analysis into understanding underlying relationships that can provide further insights.

```python
# For better represenation of our correlation results we create a heatmap
sorted_correlation_result = correlation_result.sort_values(ascending=False)
plt.figure(figsize =(10,8))
sns.heatmap(sorted_correlation_result.to_frame(), annot=True,cmap="rocket_r",fmt=".2f")
plt.show()
```

# Exploratory Analysis of Factors that Influence Students to Dropout

-  Provided that Debt is one of the major factors that influence studens' dropout from higher institutions, we selected this factor along with Student Age and Gender to come up with further insights. The results are as follows:


### 1) Student Dropout Count 

Key Insights:    
-   Almost 1/3 of the total registered students dropped out by the end of the second semester.


```python
# Student dropout count. The Non-dropout column excludes 'Enrolled students'.
plt.figure(figsize=(6,7))
sns.countplot(x="Dropout", data=df)
plt.title('Student Dropout Count')
plt.xlabel('Dropout(0=Non-dropout, 1=Dropout)')
plt.show()

```

### 2)Student Dropout by Gender

**Key Insights:**
-   A slightly higher number of female students dropped by the end of the second semester.

```python
# Count of student dropout and grouped by Gender
plt.figure(figsize=(10,12))
dropout_df=df[df["Dropout"]==1]
sns.catplot(x="Gender", data=dropout_df,kind="count", height=6)
plt.title("Student Dropout by Gender")
plt.xlabel('Gender(0=Female, 1=Male)') 
plt.show()
```

### 3) Students' Debt by Age

**Key Insights:**
-  The highest debtors among students are between 20 and 30 years old.

```python
# Distribution of number of Debtors by student age.
plt.figure(figsize=(10,7))
debtor_data=df[df['Debtor']==1]['Age at enrollment']
_=plt.hist(debtor_data,bins=5)
_=plt.xlabel('Student Ages')
_=plt.ylabel('# of Debtors')
plt.title('Debtor by Student Age')
plt.show()
```

### 4) Students' Debt by Gender

**Key Insights**:
-  Female students tend to have more debt than male students.

```python
# Number of debtors grouped by Gender.

debtor_data=df[df['Debtor']==1]
gender_counts=debtor_data['Gender'].map({0:'Female', 1:'Male'}).value_counts()
plt.figure(figsize=(10,8))
gender_counts.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('# of Debtors')
plt.title('Debtor by Gender')
plt.show
```

# Exploratory Analysis of Factors Influencing Students Reach Graduation

- The top factors influencing students towards graduation are education performance and socioeconomic aspects.    


### 1)Student Graduate Count

**Key Insights:**
-  Almost 2/3 of register students are graduate prospects.

```python
# Count of Graduted and non-Graduated students through 1st and 2nd semester. The count of Non-Graduated students excludes 
# enrolled students.
plt.figure(figsize=(6,7))
sns.countplot(x='Graduate',data=df)
plt.xlabel('Graduate(0=Non-Graduate,1=Graduate )')
plt.title("Graduated Students")
plt.show()
```

### 2) Count of Graduates by Gender

**Key Insights:**
-  Although there is double the number of registered female students compared to male students for the first and second semesters, there is more than double the number of female graduates compared to males.


```python
# Count of prospect Graduates by Gender.
plt.figure(figsize=(10,12))
Graduate_df=df[df["Graduate"]==1]
sns.catplot(x="Gender", data=Graduate_df,kind="count", height=6)
plt.xlabel('Gender(1=Male, 0=Female)') 
plt.title("Graduates by Gender")
plt.show()
```

### 3) First and Second Approved Semesters by Gender

Key Insights:

-  First Semerter: Double the number of Female Students approved the first semester over male students. While this is a considerable difference, we need to highlight that registered female students are considerably higher than males.
-  Second semester: Both female and male students see a slight decline in approved units from the First Semester.


```python
# Count of approved unit semester by Gender
fig, axes = plt.subplots(1,2, figsize=(10,8), sharey=True)
axes = axes.flatten()
plt.suptitle('')
ax1 = axes[0]
df[df['Curricular units 1st sem (approved)'] > 0]['Gender'].value_counts().plot(kind='bar', ax=ax1)
ax1.set_title('1st Semester Approved Students')
ax1.set_xlabel('Gender(0=Female,1=Male)')
ax2 = axes[1]
df[df['Curricular units 2nd sem (approved)'] > 0]['Gender'].value_counts().plot(kind='bar', ax=ax2)
ax2.set_title('2nd Semester Approved Students')
ax2.set_xlabel('Gender(0=Female,1=Male)')
plt.show()
```

### Median calculation of Semester Units Approved by Gender

```python
# Each female and male student registered for specific number of units. Below is the median number of semester units 
# approved by Gender.
Median_approved_units= df.groupby(['Gender'])[['Curricular units 1st sem (approved)','Curricular units 2nd sem (approved)']].median()
print(Median_approved_units)
```

### Visualization of Median calculation for Semerster Units Approved by Gender

**Key Insights:**
-  The use of the median is significant as it provides us with valuable insights into the central tendency of the data, enhancing our understanding of the dataset.
-  The median on Semester Units approved by Gender spans a range of 4 to 6 units.
-  Both boxplots reflected many outliers, which were visually more pronounced for female students. 


```python
# Create a boxplot to visualize a more central value of approved semester units between Male and Female students.      
fig, axes = plt.subplots(1,2, figsize=(10,8), sharey=True)
axes = axes.flatten()
plt.suptitle('')
ax1=sns.boxplot(x='Gender',y='Curricular units 1st sem (approved)',data=df,ax=axes[0])
ax1.set_title('1st Semester Approved Units')
ax1.set_xlabel('Gender(0=Female,1=Male)')
ax2=sns.boxplot(x='Gender',y='Curricular units 2nd sem (approved)', data=df,ax=axes[1])
ax2.set_xlabel('Gender(0=Female,1=Male)')
ax2.set_title('2nd Semester Approved Units')
plt.show()
```

### 4) Grades calculation for 1st and 2nd Semester by Gender

**Key Insights:**

-  Although not considerably different, both semesters have a higher median grade for female students.
-  Higher grades and performance consistency are noticeable among female students.
-  Male students had a slight decline in grades from 1st to 2nd semester.

```python
# Median of approved grades for 1st semester grouped by Gender 
plt.figure(figsize=(8,7))
sns.barplot(x="Gender", y="Curricular units 1st sem (grade)", data=df, estimator='median')
plt.title('Median for 1st Semester Grades between Males and Females')
plt.xlabel('Gender(0=Female, 1=Male)')
plt.ylabel('Median Grade')
plt.show()
```

```python
# Median of approved grades for 2nd semester grouped by Gender.
plt.figure(figsize=(8,7))
sns.barplot(x="Gender", y="Curricular units 2nd sem (grade)", data=df, estimator='median')
plt.title('Median for 2nd Semester Grades between Males and Females')
plt.xlabel('Gender(0=Female, 1=Male)')
plt.ylabel('Median Grade')
plt.show()
```

### 5) Tutition Fees up to Date

**Key Insights:**
-  The majority of female students have tuition fees paid to date.
-  About 1250 Male students are up to date with their tuition fees, leaving about 300 who still need to be, representing about 20% of all male students. 


```python
#Tuition fees paid up to date
plt.figure(figsize=(10,12))
Tuition_fees_df=df[df["Tuition fees up to date"]==1]
sns.catplot(x="Gender", data=Tuition_fees_df,kind="count",height=6)
plt.title('Tuition Fees up to Date')
plt.xlabel('Gender(0=Female, 1=Male)') 
plt.show()
```

### Summary of Insights

-  About one-third of all registered students dropped out by the end of the second semester, with female students having a    slightly higher number of dropouts than male students.
-  Debt and Age have a relatively strong influence on students' dropping out of school. Students between the ages of 20 and 30 tend to incur higher debts, and female students are slightly more in debt than males. 
-  Female students graduated at double the rate as their male counterparts. This is also reflected in the amount of females approving their curriculum units over the males. 
-  The median of approved units per semester is 4 and 6 for Males and Females, respectively.
-  Higher grades and consistency are more noticeable among female students. For male students, there was a noticeable decline in grades from first to second semester. 
-  Despite females being the group with slightly higher debt than male students, they had a lower rate of unpaid tuition fees than males. There were 300 males, which represented 20% of all registered male students, who had outstanding unpaid tuition fees for the first and second semesters.
